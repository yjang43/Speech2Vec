```python
import matplotlib.pyplot as plt

from python_speech_features import mfcc
from scipy.io import wavfile
```

```python
def locate_audio(wav, sr, from_, to):
    wav_ = wav[int(sr * from_): int(sr * to) + 1, :]
    return wav_
```

```python
sr, wav = wavfile.read('01.wav')
call = locate_audio(wav, sr, 1, 1.3)[:, 0]
spec = plt.specgram(call)
```

```python
sr, wav = wavfile.read('01.wav')
call = locate_audio(wav, sr, 1, 1.3)[:, 1]
spec = plt.specgram(call)
```

```python
sr, wav = wavfile.read('01.wav')
call = locate_audio(wav, sr, 1, 1.3)[:, 2]
spec = plt.specgram(call)
```

```python
sr, wav = wavfile.read('01.wav')
call = locate_audio(wav, sr, 1, 1.3)[:, 3]
spec = plt.specgram(call)
```

```python
feat = mfcc(call, sr, winlen=0.001, winstep=0.0005, numcep=13)
```

```python
plt.imshow(feat, aspect='auto')
```
